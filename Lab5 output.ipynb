### Exercise four
Make a bar plot of the number of tweets per day.

Hint: Use groupy() and size(). If you apply the groupby() to a column, and then apply the size() function that that, you can count the number of times the different column values exist in the data. In this case, we need the "date" part of the timestamp. Look up the pandas function groupby(), the "Time/Date Components" information at https://pandas.pydata.org/pandas-docs/stable/timeseries.html, and applying the size() function to groupby(). Finally, the plot command has an option to make different kinds of plots.

```python
df['Day/Month'] = df['DateTime'].apply(lambda x: "%d/%d" % (x.day, x.month))
(df.groupby(['Day/Month']).size()).plot(kind='bar')
df['DateTime'].groupby(df['DateTime'].dt.date).size().plot(kind='bar')
```

### Exercise five
Make a new smaller data frame containing only the tweets on the 9th. Called it df1, and print how many tweets there were on that day. Then, in a new cell, make a bar plot of the number of tweets per hour. 

```python
df1 = df[(df['Day/Month'] == '9/9')]
df1['Tweet'].count()
```

```python
df1['Hour'] = df1['DateTime'].apply(lambda x: "%2d" % (x.hour))
(df1.groupby(['Hour']).size()).plot(kind='bar')
```
